```python
text = """Hello! This is the first paragraph. It contains punctuation marks, such as commas, periods, and exclamation marks!
Do you like NLP? Yes, it's amazing: isn't it?

Here is the second paragraph; it also includes punctuation. Text preprocessing is important, because it cleans data!
Why should we do it? To improve model accuracy, of course."""
```

```python
from nltk.tokenize import word_tokenize
from gensim.models import Word2Vec
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')

tokens = word_tokenize(text.lower())
data = [tokens]
```

```python
model = Word2Vec(sentences=data, vector_size=50, window=3, min_count=1)
```

```python
# Vocabulary size
print("Vocabulary Size:", len(model.wv.index_to_key))

# Show all words
print("\nVocabulary Words:\n", model.wv.index_to_key)

# Word vector example
print("\nVector for word 'text':\n", model.wv['text'][:10])

# Another word vector
print("\nVector for word 'data':\n", model.wv['data'][:10])
```

```python
print("\nMost similar words to 'text':")
print(model.wv.most_similar('text'))
```
